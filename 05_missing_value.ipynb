### fillna

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'].fillna(100).tail()  #원본 수정X 수정된 결과값만 보여줌
```

```python
df1['age'] = df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
df1['deck'].fillna('C') #category는 이미 들어있는 데이터만 넣을 수 있음
```

```python
# df1['deck'].fillna('No Data') => 에러(카테고리 타입은 카테고리에 없는 정보는 넣을 수 없음)
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
```

```python
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치 = 남자 승객의 평균
- 여자 승객의 결측치 = 여자 승객의 평균


```python
df1 = df.copy()

c1 = df1['sex'] == 'male'
man_mean_age = df1.loc[c1, 'age'].mean()
df1.loc[c1, 'age'] = df1.loc[c1, 'age'].fillna(man_mean_age)
```

```python
c2 = df1['sex'] == 'female'
woman_mean_age = df1.loc[c2, 'age'].mean()
df1.loc[c2, 'age'] = df1.loc[c2, 'age'].fillna(woman_mean_age)
```

```python
df1['age'].mean()
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0])
```
